```python
import fitz  # PyMuPDF

def extract_text_from_pdf(pdf_file):
    text = ""
    pdf_document = fitz.open(pdf_file)

    for page_num in range(pdf_document.page_count):
        page = pdf_document.load_page(page_num)
        text += page.get_text()

    pdf_document.close()
    return text

# Specify the path to your PDF file
pdf_file_path = 'output_6.pdf'

# Extract text from the PDF
textextract = extract_text_from_pdf(pdf_file_path)

# Print the extracted text
print(textextract)
```

```python
import PyPDF2
import re

# Function to extract titles from a PDF
def extract_titles_from_pdf(pdf_file_path):
    # Initialize a list to store extracted titles
    titles = []

    # Open the PDF file
    with open(pdf_file_path, 'rb') as pdf_file:
        pdf_reader = PyPDF2.PdfReader(pdf_file)

        # Iterate through each page of the PDF
        for page_num in range(len(pdf_reader.pages)):
            page = pdf_reader.pages[page_num]
            page_text = page.extract_text()

            # Split the page text into lines
            lines = page_text.split('\n')
            # Use regular expression to find lines starting with "CHAPTER" followed by a digit
            for line in lines:
                if re.match(r'\s+\d\.\d\.\d\s+[A-Z][A-Za-z\s]+', line):
                    titles.append(line)

    return titles

# Extract titles from the PDF file (replace 'your_pdf_file.pdf' with the actual file path)
pdf_file_path = 'output_6.pdf'
extracted_titles = extract_titles_from_pdf(pdf_file_path)
SubTitle = [title for title in extracted_titles if title.count('.') <= 2]

# Print the extracted titles
# filtered_titles
SubTitle
```

```python
from nltk.corpus import wordnet as wn
def lemmatize_text (text):
    lemmatizer = WordNetLemmatizer()
    return lemmatizer.lemmatize(text)
def synonym (text):
    words = text.split()
    words = [lemmatize_text(i) for i in words]
    synonym = []
    for word in words:
        for syn in wn.synsets(word):
            for l in syn.lemmas():
                synonym.append(l.name())
    return set(synonym)
```

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.probability import FreqDist
def extract_keywords(text):
    words = word_tokenize(text)
    words = [word.lower() for word in words if word.isalpha()]
    stop_words = set(stopwords.words('english'))
    words = [word for word in words if word not in stop_words]
    fdist = FreqDist(words)
    keywords = [word for word, freq in fdist.most_common(5)]  # Adjust the number of keywords as needed
    return ', '.join(keywords)

# Add a "Keywords" column to df_sorted
df_test['Keywords'] = df_test['Content'].apply(extract_keywords)
```
